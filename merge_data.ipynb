```python
import pandas as pd
```

```python
# Load both datasets (replace with actual file paths)
crime_data_path = 'Resources/Cleaned_NYPD_Complaint_Data.csv'
housing_data_path = 'Resources/NY-House-Dataset.csv'

crime_data = pd.read_csv(crime_data_path)
housing_data = pd.read_csv(housing_data_path)

crime_data.head()
```

```python
# Drop unnecessary columns from housing_data
columns_to_drop = ['BROKERTITLE', 'ADDRESS', 'MAIN_ADDRESS', 'ADMINISTRATIVE_AREA_LEVEL_2', 'LOCALITY', 'SUBLOCALITY', 'STREET_NAME', 'LONG_NAME', 'FORMATTED_ADDRESS']
housing_data = housing_data.drop(columns=columns_to_drop)

# Display the first few rows of the updated DataFrame
housing_data.head()
```

```python

# Create a function to map state to borough names based on patterns in the STATE column
def map_state_to_borough(state):
    if "New York, NY" in state:
        return "MANHATTAN"
    elif "Staten Island, NY" in state:
        return "STATEN ISLAND"
    elif "Brooklyn, NY" in state:
        return "BROOKLYN"
    elif "Bronx, NY" in state:
        return "BRONX"
    elif "Queens, NY" in state:
        return "QUEENS"
    else:
        return "UNKNOWN" 

# Apply the function to create a new column 'BOROUGH_NAME'
housing_data['BOROUGH_NAME'] = housing_data['STATE'].apply(map_state_to_borough)

# Drop the 'STATE' column
housing_data = housing_data.drop(columns=['STATE'])

# Display the updated housing data with the new 'Borough_Name' column
housing_data.head()
```

```python
# Merge the datasets on the 'Borough_Name' column
merged_data = pd.merge(crime_data, housing_data, on='BOROUGH_NAME', how='inner')

# Display the first few rows of the merged DataFrame
merged_data.head()
```
